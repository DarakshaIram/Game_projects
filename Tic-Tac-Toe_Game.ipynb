```python
def display_game(lst):
    print(lst[0] + '||' + lst[1] + '||'+ lst[2])
    print("=  =  =")
    print(lst[3] + '||' + lst[4]+ '||' + lst[5])
    print("=  =  =")
    print(lst[6] + '||' + lst[7]+ '||' + lst[8])
```

```python
def Selection_choice():
    
    user_choice = ['x','o']
    choose = 'Wrong'
    
    while choose not in user_choice:
        choose = input("Choose a 'x' or 'o': ")
        
        if choose not in user_choice:
            print("Invalid Value")
        
        if choose in user_choice:
            return choose
```

```python
def players():
    
    selected = Selection_choice()
    player_1 = selected
    
    if selected == 'x':
        player_2 = 'o'
    else:
        player_2 = 'x'
    
    print("player_1 selection:" + player_1)
    print("player_2 selection:" + player_2)
    
    p = [player_1,player_2]
    
    return p
```

```python
def position_choice():
    
    choice='Wrong'
    acceptable_values = range(0,9)
    within_range = False
    
    while choice.isdigit() == False or within_range == False:
        choice = input("Enter any values in (0-9): ")
        
        if choice.isdigit() == False:
            print("Sorry! invalid value")
        
        if choice.isdigit() == True:
            if int(choice) in acceptable_values:
                within_range = True
            else:
                within_range = False
                print("Sorry! invalid value")
                
    position = int(choice)
    return position
```

```python
def replacement_choice(lst,position,p):
    lst[position] = p
    print(lst)
    display_game(lst)
```

```python
def line_up(lst,turn):
     
    for i in [0,1,2]:
        
        if lst[i]!= ' ' and lst[i] == lst[i+3] == lst[i+6]:
            print("You are the winner " + turn)
            return 1
        
        else:
            pass
        
    for i in [0,3,6]:
        
        if lst[i]!= ' ' and lst[i] == lst[i+1] == lst[i+2]:
            print("You are the winner " + turn)
            return 1
        
        else:
            pass
        
    if lst[0] == lst[4] == lst[8] != ' ':
        print("You Won  " + turn)
        return 1
    
    if lst[2] == lst[4] == lst[6] != ' ':
        print("You Won " + turn) 
        return 1
    
    return 0
```

```python
def final_game(lst):
    
    print("Welcome to Tic Tac Toe game")
    display_game(lst)
    p = players()
    locator=[]
    turn='p1'
        
    while len(locator) < 9:
        
        print("length of loactor")
        print( len(locator))
        
        position = position_choice()
       
        
        if position not in locator:
            
            if turn == 'p1':
                print("inif")
                replacement_choice(lst,position,p[0])
                locator.append(position)
                a = line_up(lst,turn)
                if a == 1:
                    break
                print(locator)
                turn = 'p2'
            else:
                print("first else")
                replacement_choice(lst,position,p[1])
                locator.append(position)
                a = line_up(lst,turn)
                if a == 1:
                    break
                print(locator)
                turn = 'p1'
        else:
            print("second else")
            print("Sorry! this position is already choosen. So, you choose some other position which is not choosen earlier")
            position = position_choice()
            
    if len(locator) == 9:
        print("Oops! try again :|")
    else:
        print("Congratulations !!")
        
    i =input("Do you wanna play the game again? yes or no  ")
    
    if i == 'yes':
        final_game([' ',' ',' ',' ',' ',' ',' ',' ',' '])
    else:
        print("Thank you")
```
